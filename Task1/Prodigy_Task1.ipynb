```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_394221.csv", skiprows=4)
```

```python
pop_2023 = df[['Country Name', '2023']].dropna()
```

```python
top_countries = pop_2023.sort_values(by='2023', ascending=False).head(10)
```

```python
plt.figure(figsize=(10,6))
plt.bar(top_countries['Country Name'], top_countries['2023'], color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.title("Top 10 Most Populous Countries (2023)")
plt.xlabel("Country")
plt.ylabel("Population")
plt.tight_layout()
plt.show()
```
